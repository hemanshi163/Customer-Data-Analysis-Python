```python
final_df = pd.merge(df, df1, on="CustomerID", how="right")
```

```python
final_df['TransactionDate'] = pd.to_datetime(final_df['TransactionDate'])
final_df['JoinDate'] = pd.to_datetime(final_df['JoinDate'])
```

```python
# RFM applied on final_df by using groupby and agg functions
```

```python
df_rfm= final_df.groupby("CustomerID").agg(monetary= ('TransactionAmount', 'sum'),
                                           frequency= ('TransactionDate','count'))
reference_date = final_df['TransactionDate'].max()+pd.Timedelta(days=1)
last_transaction = final_df.groupby('CustomerID')['TransactionDate'].max()
recency = (reference_date-last_transaction).dt.days
df_rfm['recency']= recency
```

```python
#Providing labels to the R_score, F_score and M_score.
```

```python
df_rfm['R_score'] = pd.qcut(df_rfm['recency'], q=5, labels=[5,4,3,2,1])
df_rfm['F_score'] = pd.qcut(df_rfm['frequency'], q=5, labels=[1,2,3,4,5])
df_rfm['M_score'] = pd.qcut(df_rfm['monetary'], q=5, labels=[1,2,3,4,5])
```

```python
df_rfm
```

```python
#As asked in the question adding new column - Segment to df_rfm (has been converted to str to get the desired output like '453')
```

```python
df_rfm['Segment'] = (df_rfm['R_score'].astype('str')+ df_rfm['F_score'].astype('str')+df_rfm['M_score'].astype('str'))
```

```python
df_rfm
```

```python
#Assigning labels to the segment using df.loc
```

```python
df_rfm['SegmentLabels'] = 'Others'    #values which will not fulfill any condition will be marked as others
```

```python
#1 Champions:
df_rfm.loc[ (df_rfm['R_score'] >=4) &
        (df_rfm['F_score'] >=4) &
        (df_rfm['M_score'] >=4),
        'SegmentLabels'] = 'Champions'

#2 Loyal Customers:
df_rfm.loc[ (df_rfm['R_score'] >=3) &
        (df_rfm['F_score'] >=4) &
        (df_rfm['M_score'] >=3),
        'SegmentLabels'] = 'Loyal Customers'

#3 Potential Loyalist
df_rfm.loc[ (df_rfm['R_score'] >=4) &
        (df_rfm['F_score'] <=3) &
        (df_rfm['M_score'] >=2),
        'SegmentLabels'] = 'Potential Loyalist'

#4 At Risk
df_rfm.loc[ (df_rfm['R_score'] <=2) &
        (df_rfm['F_score'] >=3) &
        (df_rfm['M_score'] <=2),
        'SegmentLabels'] = 'At Risk'

#5 Hibernating
df_rfm.loc[ (df_rfm['R_score'] <=2) &
        (df_rfm['F_score'] <=2) &
        (df_rfm['M_score'] <=2),
        'SegmentLabels'] = 'Hibernating'
```

```python
#Bar chart for count of cutomers in each segment using matplotlib
```

```python
import matplotlib.pyplot as plt
segment_counts = df_rfm['SegmentLabels'].value_counts()     #count of customers in each segment

plt.figure(figsize=(8,5))
plt.bar(segment_counts.index, segment_counts.values)
plt.xlabel("Segments")
plt.ylabel("Count of customers")
plt.title("Count of customers in each segment")
```

```python
#Pie chart for revenue contribution per each segment
```

```python
revenue_by_segment = df_rfm.groupby('SegmentLabels')['monetary'].sum()
labels = revenue_by_segment.index
color = ['pink','lightblue','coral','purple','yellow','magenta']

plt.pie(revenue_by_segment, labels=labels, autopct='%1.1f%%', colors= color, shadow=True)
plt.title("Revenue contribution per segment")
```

```python
#Scatter plot for recency vs monetary coloured as per each segment
```

```python
for segment in df_rfm['SegmentLabels'].unique():
    temp = df_rfm[df_rfm['SegmentLabels'] == segment]
    plt.scatter(temp['recency'], temp['monetary'], label=segment)

plt.xlabel('Recency (days since last purchase)')
plt.ylabel('Monetary (Total spend)')
plt.title('Recency vs Monetary by Customer Segment')
plt.legend()
```
